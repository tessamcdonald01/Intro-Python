```python
import matplotlib.pyplot as plt
import numpy as np
import cartopy
import cartopy.crs as ccrs
from cartopy import config
import netCDF4 as nc
import xarray as xr
import os #operating system, helps organize files
```

```python
#Plot Axes
#plotting set of axes with plate carree projection
ax = plt.axes(projection=ccrs.PlateCarree()) 
ax.coastlines(zorder=0) #putting coastlines first
ax.add_feature(cartopy.feature.LAND,facecolor='olivedrab', edgecolor='white', zorder=1) #adding land, changing face and edge colors
ax.set_extent([-100, -50, 15, 60], crs=ccrs.PlateCarree()) #changing the spatial extent we see on our map
ax.plot(-78.87,42.88, 'y*',zorder=4,transform=ccrs.PlateCarree(), markersize=20)
```

```python
#opening the netCDF file that was supplied to us by cartopy

fname  = os.path.join(config["repo_data_dir"], 'netcdf','HadISST1_SST_update.nc')
dataset = xr.open_dataset(fname,decode_times=False)
sst=dataset['sst']
sst.plot()
```

```python
#Putting the two plots together; just need to put sst on the ax that we defined earlier

ax = plt.axes(projection=ccrs.PlateCarree()) 
ax.coastlines(zorder=0) #putting coastlines first
ax.add_feature(cartopy.feature.LAND,facecolor='black', edgecolor='white', zorder=1) #adding land, changing face and edge colors
ax.set_extent([-100, -50, 15, 60], crs=ccrs.PlateCarree()) #changing the spatial extent we see on our map
ax.plot(-78.87,42.88, 'y*',zorder=4,transform=ccrs.PlateCarree(), markersize=20)

sst.plot(ax=ax)
ax.set_title(sst.long_name)
```

```python
ax = plt.axes(projection=ccrs.PlateCarree()) 
ax.coastlines(zorder=0) #putting coastlines first
ax.add_feature(cartopy.feature.LAND,facecolor='antiquewhite', zorder=1) #adding land, changing face and edge colors
ax.set_extent([-100, -50, 15, 60], crs=ccrs.PlateCarree()) #changing the spatial extent we see on our map
ax.plot(-74.92,40.17, 'g*',zorder=4,transform=ccrs.PlateCarree(), markersize=20)
ax.add_feature(cartopy.feature.STATES, edgecolor = 'tan', zorder=2)
ax.stock_img()
ax.text(-74.92,40.17, 'Langhorne', zorder = 5, color = 'white', transform=ccrs.PlateCarree(), horizontalalignment='left', verticalalignment='top')
```

```python
ax = plt.axes(projection=ccrs.LambertCylindrical()) 
ax.coastlines(zorder=0) #putting coastlines first
ax.add_feature(cartopy.feature.LAND,facecolor='antiquewhite', zorder=1) #adding land, changing face and edge colors
ax.set_extent([-100, -50, 15, 60], crs=ccrs.LambertCylindrical()) #changing the spatial extent we see on our map
ax.add_feature(cartopy.feature.STATES, edgecolor = 'tan', zorder=2)
ax.stock_img()

sst.plot(ax=ax)
ax.set_title(sst.long_name)

df = xr.DataArray(dims=("city","lat","lon"], coords={"city":[Langhorne]}, {"lat":[40.17]}, {"lon":[-74.92]}))

ax.plot(df['lon'], df['lat'], transform=ccrs.LambertCylindrical(),
        ms=8, ls='',marker='g*', zorder=4, markersize=15)
```
